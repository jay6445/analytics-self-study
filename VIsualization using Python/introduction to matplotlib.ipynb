### importing the Canada Immigration dataset

```python
df_can = pd.read_excel("Canada.xlsx", sheet_name="Canada by Citizenship", skiprows = range(20), skipfooter= 2)
```

```python
df_can
```

```python
df_can.head(8)
```

```python
df_can.info()
```

```python
df_can = df_can.rename(columns ={'OdName':'Country','AreaName':'Continent','RegName':'Region'})
```

### Dropping the first index column so country becomes the index

```python
df_can
```

```python
df_can.drop(df_can.columns[3],axis=1,inplace=True)
```

### Adding the total immigration column at the end

```python
df_can['Total'] = df_can.sum(axis=1)
```

### Setting the country as the index

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can
```

### Converting coulumn names to string

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
df_can
```

```python
years = list(map(str, range(1980,2014)))
years
```

```python
df_can.loc['Haiti',years].plot(kind ='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
df_can.loc['Pakistan',years].plot(kind='area')
```
